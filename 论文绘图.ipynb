```python
import numpy as np 
from matplotlib import pyplot as plt 
import sys 
%matplotlib inline 

sys.argv.append(r'D:\Other_All\Application_Documents\Medium\For_Work\Python\FromGit\PaddleOCR-release-2.0\output\db_mv3/train_0.log')
f = open(sys.argv[-1])
t = f.read()
f.close()
t = t.split('\n')
t = [i for i in t if ', loss:' in i ]
t = [eval(i.split('loss:')[-1].split(',')[0].strip()) for i in t]
data = np.array(t).ravel()
iter = np.arange(1,(data.size)*10+1,10)
plt.plot(iter,data)
plt.xlabel('iteration')
plt.ylabel('loss')
plt.title('(a)training')
ax = plt.gcf().add_axes([0.6,0.6,0.25,0.2])
plt.plot(iter[iter>2000],data[iter>2000])
plt.title('(b)convergence')
plt.savefig('DBNet_training_2pic.png',dpi = 200)
plt.show()
```
